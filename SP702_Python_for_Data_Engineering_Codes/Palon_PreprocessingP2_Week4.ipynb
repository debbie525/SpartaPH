```python
'''
PART 1

Using data found in sales.csv perform the following tasks:
1. Standardized the Price column
2. Identify and remove outliers in dataset

 
'''
# This is the content of sales.csv

import pandas as pd
import numpy as np
data=pd.read_csv('sales.csv')
data
```

```python
#check the initial characterization of entire dataframe

data.describe(include='all')
```

```python
'''PART 1.1 '''

# check for duplicates

data.duplicated().sum()
```

```python
'''PART 1.1'''
# remove duplicate row values
data2=data.drop_duplicates()
data2
```

```python
'''PART 1.1'''
# check for missing values

data2.isnull().sum()
```

```python
'''PART 1.1'''

#show the rows containing null values

null_columns=data2.columns[data2.isnull().any()]
data2[data2.isnull().any(axis=1)][null_columns]  

```

```python
'''PART 1.1'''

# There are multiple columns that have missing values
# Check the data and impute  the missing value in 'Price'
# I decided to keep the missing values in other columns as this will not affect my goal to standardized and remove outliers in 'Price'

new_value={'13,000': '13000'}                       #replace this data in 'Price' to correct the format
data3=data2.replace({'Price':new_value})

data3['Price']=data3['Price'].astype(float)         #convert the data type of 'Price' into float

mean_price= data3['Price'].mean()                  #get the mean 
print('Mean Price: ' + str(mean_price))

data3['Price']=data3['Price'].fillna(mean_price)   # fill-in/impute the mean price in missing value
data3.isnull().sum()                               # check if the missing value in 'Price' was filled-in


```

```python
'''PART 1.1'''

#Formula for standardization
# Xnew= (Xold-Mean)/(Standard Deviation)

data3.insert(3, 'Standardized_Price',0)
data3.Standardized_Price= (data3.Price-np.mean(data3.Price))/(np.std(data3.Price))
data3





```

```python
'''PART 1.2'''

data3.describe(include='all')



```

```python
'''PART 1.2'''
# Calculate the lower and upper limit of Price

m=np.mean(data3.Price)
s=np.std(data3.Price)
print('Mean: ' + str(m))
print('Std: ' + str(s))

ll=max( 0,m - 3*s)
ul=m + 3*s

print('Lower Limit: ' + str(ll))
print('Upper Limit: ' + str(ul))
```

```python
'''PART 1.2'''
# Identify the Outliers, print the values the out of limit range

data4=data3[data3.Price<=ll]
data4


```

```python
data5=data3[data3.Price>=ul]
data5

#Below are the outliers
```

```python
#Remove the outliers

data_cleaned1=data3[data3.Price>=ll]
data_cleaned2=data3[data3.Price<=ul]
```

```python
data_cleaned2
```

```python
# Save the cleaned data in csv file
data_cleaned2.to_csv('Palon_sales.csv', index=False)

#read the new csv file
part_1=pd.read_csv('Palon_sales.csv')
part_1
                   
```

```python
'''
Part 2:

Using the data in Handling Missing Values.csv

2.1. Handle all the missing values >provide an explanation for choosing a specific method (delete, impute or keep, if impute
why mean , median or mode?) in handling specific missing values.


'''

#this is the content of Handling Missing Values. csv

import sys
import pandas as pd
import numpy as np
data=pd.read_csv('Handling_Missing_Values.csv')
data
```

```python
'''Part 2'''

#check the initial characterization of entire dataframe

data.describe(include='all')
```

```python
'''PART 2'''
# check for missing values

data.isnull().sum()
```

```python
'''PART 2'''

#show the rows containing null values

null_columns=data.columns[data.isnull().any()]
data[data.isnull().any(axis=1)][null_columns]  
```

```python
'''PART 2'''
#delete rows with less than 5 non- NaN values, rows with less than  5 values is useless
data2=data.dropna(thresh=5)
data2
```

```python
'''PART 2'''

#fill in the missing value in  categorical column "Sex" using MODE (use Mode since this is a categorical data)

sex_value=data2['Sex'].mode()                       #get the mode of column "Sex"
print('Mode of Sex:' +  sex_value)
data2['Sex']=data2['Sex'].fillna(sex_value[0])                   # fill NaN,  
data2
 
```

```python
'''PART 2'''

#fill in the missing value in  column "Income" using MEAN (use mean since this is a numeric data)

income_value=round(data2['Income'].mean(),0)                #get the mean of column "Income"
print('Mean of Income:' +  str(income_value))
data2['Income']=data2['Income'].fillna(income_value)         # fill NaN,  
data2
```

```python
'''PART 2'''

#fill in the missing value in  column "Children" using MEAN (use mean since this is a numeric data)

children_value=round(data2['Children'].mean(),0)                #get the mean of column "Children"
print('Mean of Children:' +  str(children_value))
data2['Children']=data2['Children'].fillna(children_value)         # fill NaN,  
data2
```

```python
# Save the completed dataframe

data2.to_csv('Palon_Handling_Missing_Values.csv', index=False)

# read the new csv file
part_2=pd.read_csv('Palon_Handling_Missing_Values.csv')
part_2
```

```python
'''Part 3: Application

1. Gather data from any source (it would be better if you use your own data)
2. Preprocess the data you gathered

'''

# this is the content of Palon_Application_Raw.csv

import pandas as pd

data=pd.read_csv('Palon_Application_Raw.csv')
data
```

```python
'''Part 3'''

#check for duplicates

data.duplicated().sum()
```

```python
'''Part 3'''

#remove the 2 duplicate rows

data3=data.drop_duplicates()
data3
```

```python
'''PART 3'''
# check for missing values

data3.isnull().sum()
```

```python
'''PART 3'''

#show the rows containing null values

null_columns=data3.columns[data3.isnull().any()]
data3[data3.isnull().any(axis=1)][null_columns]  
```

```python
'''PART 3'''
#delete rows with less than 4 non- NaN values, rows with less than  5 values is useless
data3=data3.dropna(thresh=4)
data3
```

```python
# Get the Mode of Categorical Data with Missing values
# Fill-in the missing values using the resultant mode

gender_mode=data3['Gender'].mode()                      
print('Mode of Gender:' +  gender_mode)
data3['Gender']=data3['Gender'].fillna(gender_mode[0])                   

place_mode=data3['Place of Work'].mode()                      
print('Mode of Place of Work:' +  place_mode)
data3['Place of Work']=data3['Place of Work'].fillna(place_mode[0])  

transpo_mode=data3['Mode of Transportation'].mode()                      
print('Mode of Transportation:' +  transpo_mode)
data3['Mode of Transportation']=data3['Mode of Transportation'].fillna(transpo_mode[0])

data3
```

```python
'''Part 3'''

# Get the mean of numerical data with missing values
# fill in the missing values using resultant mean

age_value=round(data3['Age'].mean(),0)                
print('Mean of Age:' +  str(age_value))
data3['Age']=data3['Age'].fillna(age_value)  

time_value=round(data3['Travel Time to Work (Minutes)'].mean(),0)                
print('Mean of Travel Time:' +  str(time_value))
data3['Travel Time to Work (Minutes)']=data3['Travel Time to Work (Minutes)'].fillna(time_value)

data3
```

```python
'''Part 3'''

# Save the completed dataframe

data3.to_csv('Palon_New_Application.csv', index=False)
```
