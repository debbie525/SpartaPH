```python
# mapping using dictionary

import pandas as pd
data=pd.read_csv('New_Franchise.csv')

#create a dictionary

Location_Name={'A':'Albay', 'B':'Bulacan','C':'Cavite','D':'Davao'}
#create new column and input the data

data['Location_Name']=data['Location'].map(Location_Name)
data
```

```python
#replace a data using dictionary

#create a dictionary for data you want to replace

New_Location_Name={'Albay': 'Albania'}
new_data=data.replace({'Location_Name':New_Location_Name} )
new_data
```
