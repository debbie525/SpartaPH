```python
# Outlier Handling

import pandas as pd
import numpy as np

data=pd.read_csv('data702_outlier.csv')
print(data.describe(include='all'))
m=np.mean(data.Current)
s=np.std(data.Current)
print('Mean: ' + str(m))
print('Standard Deviation: ' + str(s))

ll=max(0, m - 3*s)
ul=m + 3*s

print('Lower Limit: ' + str(ll))
print('Upper Limit: ' + str(ul))

data_cleaned1=data[data.Current>=ll]
data_cleaned2=data[data.Current<=ul]


```

```python
#Removing Outliers
#InterQuartile Range

import pandas as pd
import numpy as np

data=pd.read_csv('data702_outlier.csv')
print(data.describe(include='all'))

IQR=np.percentile(data.Current,75) - np.percentile(data.Current,25)
print('IQR: ' +str(IQR))

ll2= max(0,np.percentile(data.Current,25)) -1.5*IQR
print('Lower Limit: ' +str(ll2))

ul2=np.percentile(data.Current,75) + 1.5*IQR
print('Upper Limit: ' + str(ul2))

data_cleaned3=data[data.Current>=ll2]
data_cleaned4=data[data.Current<=ul2]







```
