```python
# extract the review from a beach resort, https://www.kingandprince.com/

# WEB SCRAPING

from bs4 import BeautifulSoup
import urllib.request
import re  # regular expression

url = 'https://www.kingandprince.com/dining-guest-reviews.aspx'  # requests object
page = urllib.request.urlopen(url)  # connect to website
soup = BeautifulSoup(page, 'html.parser')
#print(soup)


contents = soup.find_all('p')
#print(contents)


content = []

for i in contents:
    content.append(i.getText().split('\n')[0])
#print(content)


authors = []
comments = []

i = 0
while i < len(content):
    if i==0 or i % 2 == 0:
        authors.append(content[i])
    else:
        comments.append(content[i])

    i+=1


# print(authors)
# print('---------------------------------------end---------------------------------------------')
# print('---------------------------------------end---------------------------------------------')
# print(comments)

df=pd.DataFrame(list(zip(authors,comments)), columns=['Authors', "Comments"])
                  
df = df[0:15]
df
                  
```
