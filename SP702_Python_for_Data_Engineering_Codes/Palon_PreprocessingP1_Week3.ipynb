```python
#Problem 3: Express Price in thousands
import pandas as pd   
data=pd.read_csv('sales.csv')
data=data.drop_duplicates()                                     #remove duplicate rows

new_value={'13,000': '13000'}                                   #replace this data to correct the format
data3=data.replace({'Price':new_value})

data3.dropna(subset=['Price'], inplace=True)                    #drop the rows with blank field in 'Price' column

data3['Price']=data3['Price'].astype(float)                     #convert the data type of 'Price' into float
data3.Price=data3.Price/1000                                    # Divide the 'Price' by 1000
data3.rename(columns={'Price':'Price(Thousand)'}, inplace=True) #rename the 'Price' column
data3
```

```python
'''
Problem 4: Select Name, City, State and Country data of product 1 bought by customers from united states and paid thru Amex 
or Diners. Store the selected data to empty.csv file
'''
import pandas as pd   
data=pd.read_csv('sales.csv')
data1=data[((data.Product=='Product1') & (data.Country=='United States') & (data.Payment_Type.isin(['Amex','Diners'])))] # filter the data using multiple criterias
data4=data1[['Product','Payment_Type', 'Name', 'City','State', 'Country']]       
data4.to_csv('empty.csv', index=False)

#This is to check the code for problem#4, read the contents of empty.csv file
csv_data4=pd.read_csv('empty.csv')  
csv_data4
```
